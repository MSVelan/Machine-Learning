```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
frame = cv2.imread("thor.webp")
```

```python
frame.shape
```

```python
#CV2 uses BGR instead of RGB
```

```python
plt.imshow(frame)
```

```python
rgb = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb)
```

```python
classifier = cv2.CascadeClassifier("haar_cascade_frontal_face_detection_default_xml.xml")
```

```python
faces = classifier.detectMultiScale(frame)
thor_face = faces[0]
```

```python
[x,y,w,h] = thor_face
```

```python
face = rgb[y:y+h,x:x+w]
```

```python
plt.imshow(face)
```

```python
out = cv2.rectangle(rgb,(x,y),(x+w,y+h),(0,0,255),4)
```

```python
plt.imshow(out)
```
