```python
x = np.linspace(0, 2*np.pi, 100)
```

```python
y = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x,y, label="Sin wave", color="blue", linewidth=3)
plt.plot(x,y2, label = "Cos wave",color="red", linewidth=3)
plt.title("Trig waves")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
```
